## Investigation Overview

> The goal is to check the impact of various variable on the prices of houses, specially area,neighborhood, number of bedrooms/bathrooms and style

## Dataset Overview

>  The dataset containing prices and attributes of houses like style,neighborhood area,..etc for approximately 6029 houses.


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
# load in the dataset into a pandas dataframe, print statistics
houses= pd.read_csv('house_prices.csv',encoding ='utf-8')
houses.price=houses.price/1000

houses.head()
```

> Note that the above cells have been set as "Skip"-type slides. That means
that when the notebook is rendered as http slides, those cells won't show up.

## Prices distribution

> The prices are postively skewed to the right

```python
# start with a standard-scaled plot
binsize = 100
bins = np.arange(0, houses['price'].max()+binsize, binsize)
plt.figure(figsize=[8, 5])
plt.hist(data = houses, x = 'price', bins = bins)
plt.xlabel('Price (thousand$)')

plt.ylabel('Count')
##Ignoring the tail 
plt.xlim((0,2500))
plt.title("Price destribution")
plt.show()
```

## Neighborhood houses distribution

> Looking at how houses distributed among the neighborhoods (A,B,C), we see most houses in neighborhood B

```python
houses['neighborhood'].value_counts().sort_index().plot(kind='barh' )
plt.title('Neighborhood')
#plt.ylabel('Neighborhood')
plt.xlabel('count');
sb.despine(left=True, right=True)
```

## Numerical variables correlation 

> Area and number of bedrooms/bathrooms have high correlation with price 

```python
numeric_vars = ['price', 'bedrooms', 'bathrooms', 'area']
categoric_vars = ['style', 'neighborhood']
# correlation plot
plt.figure(figsize = [8, 6])
sb.heatmap(houses[numeric_vars].corr(), annot = True, fmt = '.3f',
           cmap = 'vlag_r', center = 0)
plt.show()
```

## Numeric variables Scatter plot

> Scatter plot shows positive correlation between price and (Area,number of bedrooms ,number of bathrooms) <br>
> It also shows positive correlation between Area and (number of bathrooms,number of bedrooms)

```python
# plot matrix: 
plt.figure(figsize = [20, 20]);

g = sb.PairGrid(data = houses, vars = numeric_vars);
g = g.map_diag(plt.hist, bins = 30);
g.map_offdiag(plt.scatter);
```

## Price relation with neighborhood and style 1

> Houses prices increase in neighborhood B and also style of victorian has the highest prices 

```python
# plot matrix of numeric features against categorical features.
# can use a larger sample since there are fewer plots and they're simpler in nature.


def boxgrid(x, y, **kwargs):
    """ Quick hack for creating box plots with seaborn's PairGrid. """
    default_color = sb.color_palette()[0]
    sb.boxplot(x, y, color = default_color)

plt.figure(figsize = [10, 10])
g = sb.PairGrid(data = houses, y_vars = ['price','area'], x_vars = categoric_vars,
                height = 3, aspect = 1.5)
g.map(boxgrid)
plt.show();
```

## Price relation with neighborhood and style 2

> Houses prices increase in neighborhood B and also style of victorian has the highest prices 

```python
plt.figure(figsize = [8, 8])

# subplot 1: style vs neighborhood
plt.subplot(3, 1, 1)
sb.countplot(data = houses, x = 'style', hue = 'neighborhood', palette = 'Blues')

plt.show()
```

## Violin plot (Categorical vs Numerical) Analysis

Here are some interesting findings  
>Houses in Ranch style usually have two bathrooms <br>
> Victorian houses are more spread in area and prices <br>
> Victorian houses mostly have more bathrooms and bedrooms than any other house style <br>
> Neighborhood doesn't have a special influence in the distribution of area or number of bedrooms or bathrooms <br>

```python
# plot the categorical variables against price and area ,bedrooms and bathrooms
fig, ax = plt.subplots(ncols = 4, nrows = 2 , figsize = [15,10],constrained_layout=True)
default_color = sb.color_palette()[0]

for i in range(len(categoric_vars)):
    var = categoric_vars[i]
    sb.violinplot(data = houses, x = var, y = 'price', ax = ax[i,0],
               color = default_color)
    sb.violinplot(data = houses, x = var, y = 'area', ax = ax[i,1],
               color = default_color)
    sb.violinplot(data = houses, x = var, y = 'bedrooms', ax = ax[i,2],
               color = default_color)
    sb.violinplot(data = houses, x = var, y = 'bathrooms', ax = ax[i,3],
               color = default_color)
    
plt.show()
```

## Number of Bedrooms effect on Price

> For the same Area increase of number of bedrooms increases price 

```python
    # plotting
    #Treating bedrooms as categorical 
    bedrooms=[0,1,2,3,4,5,6,7,8]
    pd_ver = pd.__version__.split(".")
    if (int(pd_ver[0]) > 0) or (int(pd_ver[1]) >= 21): # v0.21 or later
        vclasses = pd.api.types.CategoricalDtype(ordered = True, categories = bedrooms)
        houses['bedrooms'] = houses['bedrooms'].astype(vclasses)
    else: # compatibility for v.20
        houses['bedrooms'] = houses['VClass'].astype('category', ordered = True,
                                                         categories = bedrooms)
    g = sb.FacetGrid(data = houses, col = 'bedrooms', size = 3, col_wrap = 3)
    g.map(plt.scatter, 'area', 'price', alpha = 1/5);
```
