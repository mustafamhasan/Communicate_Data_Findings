```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
# load in the dataset into a pandas dataframe, print statistics
houses= pd.read_csv('house_prices.csv',encoding ='utf-8')
houses.head()
```

```python
# high-level overview of data shape and composition
print(houses.shape)
print(houses.dtypes)
print(houses.neighborhood.value_counts())
print(houses['style'].value_counts())
```

```python
# Summary Statistics
print(houses.describe(include=['object']))
houses.describe()
```

```python
#putting the houses price in thoushand $ unit
houses.price=houses.price/1000
```

```python
# convert Neighborhood into ordered categorical types
ordinal_var_dict = {'neighborhood': ['A','B','C']}

ordered_var = pd.api.types.CategoricalDtype(ordered = True,
                                                categories = ordinal_var_dict['neighborhood'])
houses['neighborhood'] = houses['neighborhood'].astype(ordered_var)
```

## Univariate Exploration

```python
# start with a standard-scaled plot
binsize = 100
bins = np.arange(0, houses['price'].max()+binsize, binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = houses, x = 'price', bins = bins)
plt.xlabel('Price (thousand$)')
plt.show()
```

## Bivariate Exploration

> In this section, investigate relationships between pairs of variables in your
data. Make sure the variables that you cover here have been introduced in some
fashion in the previous section (univariate exploration).

```python
numeric_vars = ['price', 'bedrooms', 'bathrooms', 'area']
categoric_vars = ['style', 'neighborhood']
```

```python
# correlation plot
plt.figure(figsize = [8, 6])
sb.heatmap(houses[numeric_vars].corr(), annot = True, fmt = '.3f',
           cmap = 'vlag_r', center = 0)
plt.show()
```

> Area is the highest correlated variable with price 0.820
>  <br> Number of bedrooms has slightly more impact on price than # of bathrooms

```python
# plot matrix: 
plt.figure(figsize = [20, 20])

g = sb.PairGrid(data = houses, vars = numeric_vars)
g = g.map_diag(plt.hist, bins = 30);
g.map_offdiag(plt.scatter)
```

> Strong correaltion between price and area 
> Strong correlation between number of bedrooms and number of bathrooms


```python
# plot matrix of numeric features against categorical features.
# can use a larger sample since there are fewer plots and they're simpler in nature.


def boxgrid(x, y, **kwargs):
    """ Quick hack for creating box plots with seaborn's PairGrid. """
    default_color = sb.color_palette()[0]
    sb.boxplot(x, y, color = default_color)

plt.figure(figsize = [10, 10])
g = sb.PairGrid(data = houses, y_vars = ['price','area'], x_vars = categoric_vars,
                height = 3, aspect = 1.5)
g.map(boxgrid)
plt.show();
```

> Area destribution is not different in different neighborhood <br>
> Victorian houses has the highest prices <br>
> Neighborhood B has higher prices <br>

```python
plt.figure(figsize = [8, 8])

# subplot 1: style vs neighborhood
plt.subplot(3, 1, 1)
sb.countplot(data = houses, x = 'style', hue = 'neighborhood', palette = 'Blues')

plt.show()
```

> Neighborhood B has more houses than other areas and also has its most style of houses as victorian houses

Let's plot the violin chart for price and area against categorical variables

```python
# plot the categorical variables against price and area ,bedrooms and bathrooms
fig, ax = plt.subplots(ncols = 4, nrows = 2 , figsize = [15,10],constrained_layout=True)
default_color = sb.color_palette()[0]

for i in range(len(categoric_vars)):
    var = categoric_vars[i]
    sb.violinplot(data = houses, x = var, y = 'price', ax = ax[i,0],
               color = default_color)
    sb.violinplot(data = houses, x = var, y = 'area', ax = ax[i,1],
               color = default_color)
    sb.violinplot(data = houses, x = var, y = 'bedrooms', ax = ax[i,2],
               color = default_color)
    sb.violinplot(data = houses, x = var, y = 'bathrooms', ax = ax[i,3],
               color = default_color)
    
plt.show()
```

> Lodge houses are small in area, while victorian houses have a larger area <br>
> Neighborhood B prices are more distributed ,and the 3 neighborhood has same area distribution <br>
> Victorian has higher number of bedrooms and bathrooms and lodge has the least of them
